```python
#Combinado
df1 = pd.DataFrame({'llave':['b','b','a','c','a','a','b'],'d1':range(7)})
```

```python
df2 = pd.DataFrame({'llave':['a','b','d'],'d2':range(3)})
```

```python
df1
```

```python
df2
```

```python
merge = pd.merge(df1,df2) #Sin especificar sobre que columna 
merge
```

```python
nuevomerge = pd.merge(df1, df2, on='llave')
nuevomerge
```

```python
df3 = pd.DataFrame({'Mi_llave':['b','b','a','c','a','a','b'],'d1':range(7)})
```

```python
df4 = pd.DataFrame({'Otra_llave':['a','b','d'],'d2':range(3)})
```

```python
#
df34 = pd.merge(df3, df4, left_on='Mi_llave', right_on='Otra_llave')
```

```python
df34
```

```python
nuevol2 = pd.merge(df1, df2, how='outer')#inner, left, right
nuevol2
```
