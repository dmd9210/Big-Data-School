```python
#volumen a definir los datos pero ahora con su tipo
dtype = {'Grado': str,'Apellido Paterno': str,'Apellido Materno': str,'Nombre':str,
        'Nivel': str, 'Institución de adcripción':str,'AreaConocimiento':str}
nomColumna = ['Grado', 'APaterno', 'AMaterno', 'Nombre', 'Nivel', 'Adscripcion','AreaConocimiento']
data = pd.read_csv(arch, delimiter=',', header=1, encoding="latin1", names = nomColumna, dtype = dtype)
```
