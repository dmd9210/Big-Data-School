```python
dataT1 = pd.merge(dataA, dataB, on=['Datetime'], how='inner')
dataT1
```

```python
dataA.reset_index(drop=True) == dataB.reset_index(drop=True)
```

```python
dataT1 = dataT1.drop_duplicates(subset=['Datetime'])
dataT1
```

```python
plt.title("Values A")
dataT1['Values_x'].plot(figsize=(10,6),color='g')
plt.show();

plt.title("Values B")
dataT1['Values_y'].plot(figsize=(10,6),color='b')
plt.show();
```

```python
plt.plot(dataT1.Values_x)
plt.plot(dataT1.Values_y)
plt.plot(dataT1.Promedio)

plt.legend(["A","B","Promedio"])
plt.show();
```

```python
#Promedio
#El axis=1 significa que va sacar el promedio horizontalmente
Promedio = dataT1.mean(axis=1)
dataT1['Promedio'] = Promedio
dataT1
```

```python
#Gráfica de Promedio
plt.title('Promedio')
dataT1['Promedio'].plot(figsize=(12,6),color='r')
```

```python
#Desviación Estándar
Desviacion = dataT1.std(axis=1)
dataT1['Desviacion'] = Desviacion
dataT1
```

```python
#Gráfica de Desviación Estándar
plt.title('Desviacion')
dataT1['Desviacion'].plot(figsize=(10,6),kind='kde',color='r')
```

```python
#Varianza
Varianza = dataT1.var(axis=1)
dataT1['Varianza'] = Varianza
dataT1
```

```python
#Gráfica de Varianza
plt.title('Varianza')
dataT1['Varianza'].plot(figsize=(10,6),kind='kde',color='r')
```

```python
#Covarianza
Covarianza = dataT1.cov()['Values_x']
dataT1['Covarianza'] = Covarianza
dataT1
```

```python
#Gráfica de Covarianza
```
