<h2 style='text-align: center;'>Usamos UPDATE & UPSERT  de nuestra base de datos</h2>

```python
#Este sirve para poder utilizar los ids de la base de datos.
from bson.objectid import ObjectId
```

```python
""" 
Upsert es la combinación realiza dos funciones que son update e insert a la tabla.Tambien recordar 
para buscar un id especifico usamos el ObjectId.
""""
collection.find_one_and_update(
{"_id":ObjectId ('626c94762e5c29bdd6386cb0')},
    {
    "$set":{
        "class":"First"
        }
    },
    upsert=True
)
```

```python
collection.find_one_and_update(
{"_id":ObjectId ('626c94762e5c29bdd6386cb1')},
    {
    "$set":{
        "fare":150,
        "deck":"A"
        }
    },
    upsert=True
)
```

```python
collection.find_one_and_update(
{"_id":ObjectId ('626c94762e5c29bdd6386cba')},
    {
    "$set":{
        "age": 65,
        "embark_town":"Cherbourg",
        "alive":"no"
        }
    },
    upsert=True
)
```

<h2 style='text-align: center;'>Creamos VISTAS para la base de datos</h2>

```python
#Con el siguiente aggregate mostramos a los que sobrevivieron y si quedaron solo o no
resultado = db.titanic.aggregate(
[
    {"$group":
        { "_id":{"survived":"$survived",
        "alone": "$alone"},
        "num_surv": {"$sum":1}
        }
    }
]
)

for i in resultado:
    print(i)
```

```python
#Agrupamos por los que estan vivos.
resultado = db.titanic.aggregate(
[
    {"$group":
        { "_id":{"alive":"$alive",
        },
        "num_alive": {"$sum":1}
        }
    }
]
)

for i in resultado:
    print(i)
```

```python
#Agrupamos los supervivientes por clase
resultado = db.titanic.aggregate(
[
    {"$group":
        { "_id":{"survived":"$survived",
        "class": "$class"},
        "num_surv": {"$sum":1}
        }
    }
]
)

for i in resultado:
    print(i)
```
